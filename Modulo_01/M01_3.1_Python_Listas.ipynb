## Estructuras de datos

Son medios para almacenar datos juntos, contienen una gran cantidad de otros objetos dentro de ellos. A veces, las estructuras de datos también se denominan tipos de datos compuestos.  
Se pueden clasificar en términos generales como estructuras de datos mutables o inmutables. Como sugiere el nombre, estructuras de datos mutables.
se pueden cambiar una vez definidos. Las estructuras de datos inmutables no se pueden cambiar una vez que
han sido definidos.

Hablaremos de las siguientes estructuras de datos:  
- **List**: una colección de objetos arbitrarios; muchos métodos disponibles (MUTABLE)     
- **Dict**: un objeto de almacenamiento clave-valor (MUTABLE)  
       
También hay un par de otras estructuras de datos. Hablaremos de esto más adelante en el bootcamp.
   
- **Set**: un objeto de colección desordenado para otros objetos únicos (MUTABLE)   
- **Tuple**: una colección de objetos arbitrarios; sólo unos pocos métodos disponibles (INMUTABLE)

## Tipo de datos versus estructuras de datos:

El tipo de datos no se puede reducir, mientras que una estructura de datos sí, ya que consta de múltiples campos de datos diferentes (analogía: un tipo de datos es un átomo, mientras que las estructuras de datos son moléculas).  
Todo tipo de datos es una estructura de datos, pero no toda estructura de datos es un tipo de datos.  
Comprenderá mejor la diferencia entre los dos después de que veamos algunos ejemplos de estructuras de datos.

## Listas

**Las listas** se utilizan para almacenar varios elementos de datos en una sola variable. Siempre están encapsulados en
_corchetes_. Hay direcciones asignadas a cada valor de una lista, que se conocen como índices.   
Python sigue una indexación basada en cero, lo que significa que la posición del índice comienza con un cero y secuencialmente
incrementándolo hasta que todos los elementos de la lista.

```python
example1 = [1, 2, 3, 4]
example2 = ['a', 'b', 'c']
example3 = [1 , 'a', True]
```

**Notas**:
 - Las listas siempre están encerradas entre corchetes.
 - Los diferentes elementos de una lista están separados por comas.

**Extraer múltiples elementos de la lista - Slicing**

```python
x[:]           # regresa todo lo que esta dentro de la lsita
x[:stop]       # regresa los elementos desde el inicio hasta el elemento que este ANTES del elemento con el índice "stop"
x[start:]      # regresa desde el indice en start hasta el ultimo elemento de la lista
x[start:stop]  # regresa los elementos que estan entre los indice de start y stop
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
```

```python
x[:]
```

```python
x[:3]
```

```python
x[3:]
```

```python
x[3:5]
```

### Métodos/Funciones utilizadas en Listas

Aquí hay algunas funciones predefinidas en Python que se pueden usar en listas:  
- **función de longitud: len()**

```python
x = [12, 43, 4, 1, 6, 343, 10]
len(x)  # Numero de elementos de la lista x
```

**Nota:**
Dado que Python tiene una indexación basada en cero, el último elemento de la lista tiene un índice de 'longitud-1'

```python
x[6]
```

```python
# Último elemento de la lista
x[len(x)- 1]
```

```python
x = [1, 1.1, 23, 5.3, 5, 8.3, 'hello', True]
len(x)
```

- **Función de índice**: Se utiliza para encontrar el índice de un elemento presente en una lista.

```python
x = [1, 1.1, 23, 5.3, 5, 8.3, 'hello', True]
```

```python
x.index('hello')
```

```python
x.index(8.3)
```
